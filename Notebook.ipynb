```python
from TS.Processing import data_yf,plot_graph, plot_segm
import matplotlib.pyplot as plt
import pandas as pd
```

```python
label=['Top 10 Tech Company', 'Top 10 Health Company', 'Top 10 Energy Company']
```

```python
def data(companies,start_date,end_date):
    numbers = list(map(str, companies.split()))
    data=data_yf(numbers,start_date,end_date)
    if int(len(data['Close'].index)/10)>=10:
            precision_recommande=int(len(data['Close'].index)/10)
    else :
        precision_recommande=10
    precision=int(input("Précision des segments ({} recomandé)) : ".format(precision_recommande)))
    return data,precision
```

```python
def dataf(companies,start_date,end_date):
    comb_data=data(companies,start_date,end_date)
    combined_data=comb_data[0]['Adj Close'].mean(axis=1)
    combined_data_prec=comb_data[1]
    combined_data=pd.DataFrame(combined_data)
    combined_data.rename(columns={0 :'Close'}, inplace=True )
    return combined_data,combined_data_prec
```

```python
def dataf_nomeans(companies,start_date,end_date):
    comb_data=data(companies,start_date,end_date)
    combined_data=comb_data[0]['Adj Close']
    combined_data_prec=comb_data[1]
    combined_data=pd.DataFrame(combined_data)
    combined_data.rename(columns={0 :'Close'}, inplace=True )
    return combined_data,combined_data_prec
```

```python
def plot_nomeans(companies,start_date,end_date):
    plt.plot(dataf_nomeans(companies,start_date,end_date)[0].index,dataf_nomeans(companies,start_date,end_date)[0],label=str(companies))
    plt.xlabel('Date')
    plt.ylabel('Prix action')
    plt.gcf().autofmt_xdate()
    plt.legend()
```

```python
def plot(companies,start_date,end_date,i):
    plt.plot(dataf(companies,start_date,end_date)[0].index,dataf(companies,start_date,end_date)[0],label=label[i])
    plt.xlabel('Date')
    plt.ylabel('Prix action')
    plt.gcf().autofmt_xdate()
    plt.legend()
    # plot_segm(dataf(companies)[0],dataf(companies)[1])
#adding_image(plt.imread("logo hadamard stage.jfif"),2021)
```

```python
companies = input('liste entreprise : ')
start_date=input('Date de début (XXXX-XX-XX) : ')
end_date=input('Date de fin (XXXX-XX-XX) : ')
numbers = list(map(str, companies.split()))
i=0
for x in numbers:
    plot_nomeans(x,start_date,end_date)
plt.show(loc='best')
```

```python
start_date=input('Date de début (XXXX-XX-XX) : ')
end_date=input('Date de fin (XXXX-XX-XX) : ')
for i in range(3):
    companies = input('liste entreprise : ')
    plot(companies,start_date,end_date,i)
plt.show()
```

```python
company='AAPL'
start_date='2022-01-01'
end_date='2023-01-01'
data=data_yf(company,start_date,end_date)
fig, ax = plt.subplots()

special_date = pd.Timestamp('2022-04-25')
special_value = data[data.index==special_date]['Close']  # Exemple de valeur, remplacez par la valeur réelle à cette date

ax.plot(data.index,data['Close'],color="#C0C0C0")
plot_segm(data,7)

# Ajouter le point spécial
ax.scatter(special_date, special_value, color='Black', zorder=2)

# Ajouter une annotation avec une flèche
ax.annotate('Rachat de Twitter', xy=(special_date, special_value), xytext=(special_date - pd.Timedelta(days=20), special_value-20),
            arrowprops=dict(facecolor='white', shrink=0.01, width=2, headwidth=8),
            bbox=dict(boxstyle="round,pad=0.3", facecolor="White", edgecolor="black", linewidth=2),
            horizontalalignment='right', verticalalignment='top')

# Ajuster le format des dates sur l'axe des x pour plus de clarté
ax.xaxis.set_major_locator(plt.MaxNLocator(12))
fig.autofmt_xdate()

plt.show()
```
